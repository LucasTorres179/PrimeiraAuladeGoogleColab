# Tabela Verdade

and

```python
print(10 > 2 and 4 < 6)     # V and V - V
print(10 > 2 and 4 > 6)     # V and F - F
print(10 < 2 and 4 < 6)     # V and F - F
print(10 < 2 and 4 > 6)     # V and F - F
```

Or

```python
print(10 > 2 or 4 < 6)  # V or V - V
print(10 > 2 or 4 > 6)  # V or F - V
print(10 < 2 or 4 < 6)  # V or F - V
print(10 < 2 or 4 > 6)  # F or F - F
```

Not

```python
not 10 > 2  # V - F
not 10 < 2  # F - V
```

# Condicional

```python
# if
if 5 < 10:
  print('verdadeiro')
else:
    print('falso')
```

```python
# fi/else
num1 = 10
num2 = 5
if num1 > num2:
  print ( 'O num1 é o maior número')
else:
    print ('O num2 é o maior número')
```

```python
# If/Elif/else
num1 = 10
num2 = 5
if num1 > num2:
  print ( 'O num 1 é o maior número')
elif num1 < num2:
    print ('O num2 é o maior número')
else:
      print ('Os dois número são iguais')
```

#  Condicional com operadores lógicos

```python
num1 = 10
num2 = 5
num3 = 2
if (num1 > num2) and (num1 > num3):
  print ('O num1 é o maior número')
else:
  print ('O  num1 não é o maior número')
```

```python
num1 = 299
num2 = 345
num3 = 1000
if (num1 > num2) and (num1 > num3):
  print ('O num1 é o maior número')
elif (num1 < num2) and (num2 > num3):
  print ('O num2 é o maior número')
else:
  print ('O num3 é o maior número')
```
